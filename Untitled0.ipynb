## Data cleaning

### Subtask:
Clean the mobile sales data by handling missing values and removing duplicates.


**Reasoning**:
Clean the data by handling missing values, removing duplicates, and ensuring data type consistency.



```python
# Handling missing values
# For 'Core Specification' and 'SSD', remove rows with missing values due to high percentage.
df_sales.dropna(subset=['Core Specification', 'SSD'], inplace=True)

# Check for missing values in other columns and impute if necessary
for col in df_sales.columns:
    if df_sales[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df_sales[col]):
            df_sales[col].fillna(df_sales[col].median(), inplace=True)  # Impute with median for numerical columns
        else:
            df_sales[col].fillna(df_sales[col].mode()[0], inplace=True)  # Impute with mode for categorical columns


# Remove duplicates
df_sales.drop_duplicates(inplace=True)

# Ensure data type consistency
df_sales['Price'] = pd.to_numeric(df_sales['Price'], errors='coerce')
df_sales['Quantity Sold'] = pd.to_numeric(df_sales['Quantity Sold'], errors='coerce')
df_sales['Inward Date'] = pd.to_datetime(df_sales['Inward Date'], errors='coerce')
df_sales['Dispatch Date'] = pd.to_datetime(df_sales['Dispatch Date'], errors='coerce')

# Outlier detection and handling (optional)
# Example using IQR for 'Price'
Q1 = df_sales['Price'].quantile(0.25)
Q3 = df_sales['Price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_sales = df_sales[(df_sales['Price'] >= lower_bound) & (df_sales['Price'] <= upper_bound)]

display(df_sales.head())
print(df_sales.info())
```

## Data wrangling

### Subtask:
Prepare the data for analysis and modeling by creating new features.


**Reasoning**:
Create new features based on sales aggregation, trends, and categorical features.  Handle data types as needed.



```python
# Aggregate Sales
df_sales['Inward Date'] = pd.to_datetime(df_sales['Inward Date'])
df_sales['Monthly_Sales'] = df_sales.groupby(df_sales['Inward Date'].dt.to_period('M'))['Quantity Sold'].transform('sum')
df_sales['Quarterly_Sales'] = df_sales.groupby(df_sales['Inward Date'].dt.to_period('Q'))['Quantity Sold'].transform('sum')
df_sales['Yearly_Sales'] = df_sales.groupby(df_sales['Inward Date'].dt.to_period('Y'))['Quantity Sold'].transform('sum')

# Sales Trends: Month-over-month growth rate
df_sales['Month_Over_Month_Growth'] = df_sales.groupby('Brand')['Quantity Sold'].pct_change()

# Sales Trends: Moving averages (3-month and 6-month)
df_sales['3_Month_MA'] = df_sales['Quantity Sold'].rolling(window=3, min_periods=1).mean()
df_sales['6_Month_MA'] = df_sales['Quantity Sold'].rolling(window=6, min_periods=1).mean()


# Categorical Feature Engineering: Sales by region
df_sales['Regional_Sales'] = df_sales.groupby('Region')['Quantity Sold'].transform('sum')


display(df_sales.head())
```

## Data analysis

### Subtask:
Analyze the prepared sales data to identify trends, seasonality, and potential outliers.


**Reasoning**:
Analyze the time series data, investigate relationships between variables, detect outliers, and identify key drivers of sales.

## Feature engineering

### Subtask:
Prepare the data for time series forecasting by creating relevant features.


**Reasoning**:
Create lagged features, rolling statistics, and time-based features from the 'df_sales' DataFrame. Handle missing values generated by these operations.



```python
# 1. Lagged Features
for lag in [1, 2, 3, 6, 12]:
    df_sales[f'Quantity_Sold_Lag_{lag}'] = df_sales.groupby('Brand')['Quantity Sold'].shift(lag)

# 2. Rolling Statistics
df_sales['Rolling_Mean_3M'] = df_sales['Quantity Sold'].rolling(window=3, min_periods=1).mean()
df_sales['Rolling_Std_3M'] = df_sales['Quantity Sold'].rolling(window=3, min_periods=1).std()
df_sales['Rolling_Mean_6M'] = df_sales['Quantity Sold'].rolling(window=6, min_periods=1).mean()
df_sales['Rolling_Std_6M'] = df_sales['Quantity Sold'].rolling(window=6, min_periods=1).std()

# 3. Time-Based Features
df_sales['Month'] = df_sales['Inward Date'].dt.month
df_sales['Quarter'] = df_sales['Inward Date'].dt.quarter
df_sales['Year'] = df_sales['Inward Date'].dt.year
df_sales['DayOfWeek'] = df_sales['Inward Date'].dt.dayofweek

# 4. Handle Missing Values
df_sales.dropna(inplace=True)

display(df_sales.head())
```

## Data visualization

### Subtask:
Create interactive and futuristic visualizations of the sales forecasts.


**Reasoning**:
Create interactive visualizations of the sales forecasts using Plotly, and highlight specific aspects of the forecast.



```python
import pandas as pd
import plotly.express as px
import pickle
from prophet.plot import plot_plotly, plot_components_plotly

# Load the trained Prophet model
with open('prophet_model.pkl', 'rb') as f:
    model = pickle.load(f)

# Create a future dataframe extending to 2026
future = model.make_future_dataframe(periods=365)

# Generate sales forecasts
forecast = model.predict(future)

# Interactive forecast plot
fig1 = plot_plotly(model, forecast)
fig1.update_layout(title_text='Interactive Sales Forecast')
fig1.show()


# Interactive forecast components plot
fig2 = plot_components_plotly(model, forecast)
fig2.update_layout(title_text='Interactive Forecast Components')
fig2.show()


# Additional visualization: Trend
fig3 = px.line(forecast, x='ds', y='trend', title='Sales Trend')
fig3.show()


# Additional visualization: Seasonality
fig4 = px.line(forecast, x='ds', y='yearly', title='Yearly Seasonality')
fig4.show()
```
